Set `MODE = "SMOKE" | "DEV" | "PAPER"`. Run: `.venv/bin/python -m jupyter nbconvert --execute --inplace fig1_convergence_sigma.ipynb`


```python
import os
from pathlib import Path
for v in ("OPENBLAS_NUM_THREADS", "OMP_NUM_THREADS", "MKL_NUM_THREADS"):
    os.environ.setdefault(v, "1")

MODE = "SMOKE"  # SMOKE (~2 min) | DEV | PAPER

from logit_graph.experiments import PRESETS, run_sigma_sweep, plot_convergence_sigma

OUT_DIR = (Path("..") / ".." / "images" / "correction_paper").resolve()
OUT_DIR.mkdir(parents=True, exist_ok=True)

cfg = PRESETS[MODE]["sigma"]
print(f"MODE={MODE}, n={cfg.n_values}, reps={cfg.n_reps}")

df = run_sigma_sweep(cfg, OUT_DIR, use_cache=(MODE != "SMOKE"))
plot_convergence_sigma(df, OUT_DIR / "convergence_sigma.png")
print("Saved", OUT_DIR / "convergence_sigma.png")
df.tail()
```
