# E5 — AIC $d$ selection (`fig:aic_d`)

Uses fixed package: incremental features + penalized ensemble AIC.

Set `MODE = "SMOKE" | "DEV" | "PAPER"`.


```python
import os
from pathlib import Path
for v in ("OPENBLAS_NUM_THREADS", "OMP_NUM_THREADS", "MKL_NUM_THREADS"):
    os.environ.setdefault(v, "1")

MODE = "SMOKE"

from logit_graph.experiments import PRESETS, run_aic_d_sweep, plot_aic_confusion

OUT_DIR = (Path("..") / ".." / "images" / "correction_paper").resolve()
OUT_DIR.mkdir(parents=True, exist_ok=True)

cfg = PRESETS[MODE]["aic"]
print(f"MODE={MODE}, n={cfg.n_sizes}, runs={cfg.n_runs}, M={cfg.m_ensemble}")

_, conf = run_aic_d_sweep(cfg, OUT_DIR, use_cache=(MODE != "SMOKE"))
plot_aic_confusion(conf, cfg.d_true_values, OUT_DIR / "aic_d_confusion_n_sweep.png")
print("Saved", OUT_DIR / "aic_d_confusion_n_sweep.png")
```
